###  벡터 내적 (Dot Product)

📖 1단계: 내적이 뭔가요?
🧠 가장 간단한 정의

내적 = 같은 위치끼리 곱하고 다 더하기

🔢 수학 공식

2차원 벡터의 내적:
$$\vec{A} \cdot \vec{B} = A_1 \times B_1 + A_2 \times B_2$$
일반적인 n차원 벡터의 내적:
$$\vec{A} \cdot \vec{B} = \sum_{i=1}^{n} A_i \times B_i = A_1B_1 + A_2B_2 + ... + A_nB_n$$


```python
import numpy as np

print("🔢 내적 계산 - 아주 간단한 예시")
print("="*40)

# 아주 간단한 벡터부터
A = np.array([2, 3])
B = np.array([4, 1])

print(f"벡터 A = {A}")
print(f"벡터 B = {B}")
print()
print("내적 계산:")
print(f"A·B = A₁×B₁ + A₂×B₂")
print(f"A·B = {A[0]}×{B[0]} + {A[1]}×{B[1]}")
print(f"A·B = {A[0]*B[0]} + {A[1]*B[1]}")
print(f"A·B = {A[0]*B[0] + A[1]*B[1]}")

# 넘파이로 확인
numpy_result = np.dot(A, B)
print(f"\nNumPy 확인: {numpy_result}")
```

```python
import matplotlib.pyplot as plt

# 간단한 벡터 시각화
fig, ax = plt.subplots(figsize=(8, 6))

# 간단한 벡터들
A = np.array([2, 3])
B = np.array([4, 1])

# 벡터 그리기
ax.arrow(0, 0, A[0], A[1], head_width=0.1, head_length=0.1, 
         fc='red', ec='red', linewidth=2, label=f'A = ({A[0]}, {A[1]})')
ax.arrow(0, 0, B[0], B[1], head_width=0.1, head_length=0.1, 
         fc='blue', ec='blue', linewidth=2, label=f'B = ({B[0]}, {B[1]})')

# 원점과 끝점 표시
ax.plot(0, 0, 'ko', markersize=6)
ax.plot(A[0], A[1], 'ro', markersize=8)
ax.plot(B[0], B[1], 'bo', markersize=8)

# 좌표 표시
ax.text(A[0]+0.1, A[1]+0.1, f'A({A[0]}, {A[1]})', fontsize=10)
ax.text(B[0]+0.1, B[1]+0.1, f'B({B[0]}, {B[1]})', fontsize=10)

ax.set_xlim(-0.5, 5)
ax.set_ylim(-0.5, 4)
ax.grid(True, alpha=0.3)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('간단한 벡터들', fontsize=14)
ax.legend()

plt.tight_layout()
plt.show()

# 내적 계산 단계별 설명
dot_product = A[0]*B[0] + A[1]*B[1]
print(f"📊 그래프에서 본 벡터들의 내적:")
print(f"A·B = {A[0]}×{B[0]} + {A[1]}×{B[1]} = {dot_product}")
```

```python
print("🎯 다양한 벡터들의 내적 비교")
print("="*50)

# 경우 1: 같은 방향
A1 = np.array([2, 1])
B1 = np.array([4, 2])  # A1의 2배
dot1 = np.dot(A1, B1)
print(f"경우 1 - 같은 방향:")
print(f"A = {A1}, B = {B1}")
print(f"내적 = {A1[0]}×{B1[0]} + {A1[1]}×{B1[1]} = {dot1}")

# 경우 2: 수직
A2 = np.array([1, 0])
B2 = np.array([0, 1])
dot2 = np.dot(A2, B2)
print(f"\n경우 2 - 수직:")
print(f"A = {A2}, B = {B2}")
print(f"내적 = {A2[0]}×{B2[0]} + {A2[1]}×{B2[1]} = {dot2}")

# 경우 3: 반대 방향
A3 = np.array([2, 1])
B3 = np.array([-2, -1])
dot3 = np.dot(A3, B3)
print(f"\n경우 3 - 반대 방향:")
print(f"A = {A3}, B = {B3}")
print(f"내적 = {A3[0]}×{B3[0]} + {A3[1]}×{B3[1]} = {dot3}")

print(f"\n💡 패턴 발견:")
print(f"같은 방향 → 내적 > 0 ({dot1})")
print(f"수직 → 내적 = 0 ({dot2})")
print(f"반대 방향 → 내적 < 0 ({dot3})")
```

```python
print("⚡ 포켓몬으로 내적 연습")
print("="*40)

# 포켓몬 능력치 벡터
pikachu = np.array([55, 40])    # [공격력, 방어력]
raichu = np.array([90, 55])

print(f"피카츄 = {pikachu} (공격력: {pikachu[0]}, 방어력: {pikachu[1]})")
print(f"라이츄 = {raichu} (공격력: {raichu[0]}, 방어력: {raichu[1]})")
print()

print("단계별 내적 계산:")
print("1단계: 같은 위치끼리 곱하기")
mult1 = pikachu[0] * raichu[0]
mult2 = pikachu[1] * raichu[1]
print(f"   공격력끼리: {pikachu[0]} × {raichu[0]} = {mult1}")
print(f"   방어력끼리: {pikachu[1]} × {raichu[1]} = {mult2}")

print("\n2단계: 곱한 값들을 모두 더하기")
total = mult1 + mult2
print(f"   {mult1} + {mult2} = {total}")

print(f"\n결과: 피카츄·라이츄 = {total}")

# 검증
numpy_result = np.dot(pikachu, raichu)
print(f"NumPy 검증: {numpy_result}")
print(f"맞나요? {total == numpy_result}")
```

### 벡터 정규화(Normalizaiton)


📖 정규화가 뭔가요?
🧠 직관적 이해
정규화 = 모든 벡터를 같은 길이(1)로 만들어서 "방향"만 비교하기
🔢 수학 공식
벡터의 크기(길이, 노름):
$$||\vec{v}|| = \sqrt{v_1^2 + v_2^2 + ... + v_n^2} = \sqrt{\sum_{i=1}^{n} v_i^2}$$
정규화 공식:
$$\hat{v} = \frac{\vec{v}}{||\vec{v}||} = \frac{\vec{v}}{\sqrt{\sum_{i=1}^{n} v_i^2}}$$
정규화된 벡터의 길이는 항상 1!


```python
import numpy as np

print("📏 벡터 정규화 - 기초부터")
print("="*40)

# 간단한 벡터
A = np.array([3, 4])

print(f"원본 벡터 A = {A}")
print()

# 1단계: 벡터의 길이 구하기
print("1단계: 벡터의 길이 구하기")
print(f"||A|| = √(3² + 4²)")
print(f"||A|| = √(9 + 16)")
print(f"||A|| = √25")
print(f"||A|| = 5")

# 넘파이로 확인
length = np.linalg.norm(A)
print(f"NumPy 확인: {length}")

# 2단계: 벡터를 길이로 나누기
print(f"\n2단계: 벡터를 길이로 나누기")
normalized_A = A / length
print(f"정규화된 A = {A} / {length}")
print(f"정규화된 A = [{A[0]/length:.3f}, {A[1]/length:.3f}]")
print(f"정규화된 A = {normalized_A}")

# 3단계: 정규화된 벡터의 길이 확인
new_length = np.linalg.norm(normalized_A)
print(f"\n3단계: 정규화된 벡터의 길이")
print(f"||정규화된 A|| = {new_length:.10f}")
print("✅ 길이가 1이 되었어요!")
```

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# 왼쪽: 원본 벡터들 (길이 다름)
vectors = np.array([[3, 4], [6, 2], [1, 3], [5, 1]])
colors = ['red', 'blue', 'green', 'orange']
labels = ['A(3,4)', 'B(6,2)', 'C(1,3)', 'D(5,1)']

for i, (vec, color, label) in enumerate(zip(vectors, colors, labels)):
    ax1.arrow(0, 0, vec[0], vec[1], 
              head_width=0.2, head_length=0.3, 
              fc=color, ec=color, linewidth=2, alpha=0.7)
    length = np.linalg.norm(vec)
    ax1.text(vec[0]*1.1, vec[1]*1.1, f'{label}\n길이:{length:.1f}', 
             fontsize=9, ha='center')

ax1.set_xlim(0, 7)
ax1.set_ylim(0, 5)
ax1.grid(True, alpha=0.3)
ax1.set_title('원본 벡터들 (길이가 다름)', fontsize=12, fontweight='bold')
ax1.set_aspect('equal')

# 오른쪽: 정규화된 벡터들 (길이 모두 1)
for i, (vec, color, label) in enumerate(zip(vectors, colors, labels)):
    normalized = vec / np.linalg.norm(vec)
    ax2.arrow(0, 0, normalized[0], normalized[1], 
              head_width=0.05, head_length=0.06, 
              fc=color, ec=color, linewidth=2, alpha=0.7)
    ax2.text(normalized[0]*1.2, normalized[1]*1.2, f'{label}\n길이:1.0', 
             fontsize=9, ha='center')

# 단위원 그리기
circle = plt.Circle((0, 0), 1, fill=False, color='black', linestyle='--', alpha=0.5)
ax2.add_patch(circle)

ax2.set_xlim(-1.5, 1.5)
ax2.set_ylim(-1.5, 1.5)
ax2.grid(True, alpha=0.3)
ax2.set_title('정규화된 벡터들 (길이 모두 1)', fontsize=12, fontweight='bold')
ax2.set_aspect('equal')

plt.tight_layout()
plt.show()

print("💡 정규화의 핵심:")
print("- 원본: 벡터들의 길이가 제각각")
print("- 정규화 후: 모든 벡터의 길이가 1")
print("- 효과: 크기 무시하고 '방향'만 비교 가능!")
```

```python
print("⚡ 포켓몬 능력치 정규화")
print("="*40)

# 포켓몬 능력치
pikachu = np.array([55, 40])
raichu = np.array([90, 55])

print(f"피카츄 원본: {pikachu}")
print(f"라이츄 원본: {raichu}")

# 각각의 길이 계산
pikachu_length = np.sqrt(pikachu[0]**2 + pikachu[1]**2)
raichu_length = np.sqrt(raichu[0]**2 + raichu[1]**2)

print(f"\n길이 계산:")
print(f"피카츄 길이: √({pikachu[0]}² + {pikachu[1]}²) = √{pikachu[0]**2 + pikachu[1]**2} = {pikachu_length:.2f}")
print(f"라이츄 길이: √({raichu[0]}² + {raichu[1]}²) = √{raichu[0]**2 + raichu[1]**2} = {raichu_length:.2f}")

# 정규화
pikachu_normalized = pikachu / pikachu_length
raichu_normalized = raichu / raichu_length

print(f"\n정규화 결과:")
print(f"피카츄 정규화: [{pikachu_normalized[0]:.3f}, {pikachu_normalized[1]:.3f}]")
print(f"라이츄 정규화: [{raichu_normalized[0]:.3f}, {raichu_normalized[1]:.3f}]")

# 길이 확인
print(f"\n정규화 후 길이 확인:")
print(f"피카츄 길이: {np.linalg.norm(pikachu_normalized):.10f}")
print(f"라이츄 길이: {np.linalg.norm(raichu_normalized):.10f}")
print("✅ 둘 다 길이가 1!")

# 넘파이 함수로 간단히
pikachu_norm_numpy = pikachu / np.linalg.norm(pikachu)
print(f"\nNumPy로 간단히: {pikachu_norm_numpy}")
```

```python
print("🔍 정규화 전 vs 후 비교")
print("="*40)

# 원본 내적
original_dot = np.dot(pikachu, raichu)
print(f"원본 내적: {original_dot}")

# 정규화 후 내적
normalized_dot = np.dot(pikachu_normalized, raichu_normalized)
print(f"정규화 후 내적: {normalized_dot:.6f}")

print(f"\n💡 차이점:")
print(f"원본 내적: 크기와 방향 모두 영향")
print(f"정규화 후 내적: 방향만 영향 (= 코사인 유사도!)")

print(f"\n🎯 정규화가 중요한 이유:")
print("1. 벡터 크기에 관계없이 방향만 비교")
print("2. 코사인 유사도 계산이 쉬워짐")
print("3. 머신러닝에서 스케일 차이 제거")
print("4. 비교 기준이 일관됨")
```
