```python
import pandas as pd

#포켓몬 데이터 불러오기
df = pd.read_csv('../1일차/data/pokemon.csv')
```

```python
print("🔍 데이터 확인:")
print(f"총 포켓몬 수: {len(df)}")
print(f"컬럼들: {df.columns.tolist()}")
```

```python
# "Pikachu"가 이름에 들어간 포켓몬 찾기
pikachu_data = df[df['name'].str.contains('Pikachu', na=False)]

print("⚡ 피카츄 데이터:")
print(pikachu_data)
```

### 피카츄 능력치 찍어보기

```python
pikachu = df[df['name'] == 'Pikachu']
print(f"방법 1 결과: {len(pikachu)}마리")
```

```python
# 위에서 피카츄를 찾았다면 계속 진행
if len(pikachu) > 0:
    # 피카츄 선택 (첫 번째 피카츄)
    my_pikachu = pikachu.iloc[0]
    
    print("⚡ 피카츄 능력치:")
    print(f"이름: {my_pikachu['name']}")
    print(f"스피드: {my_pikachu['speed']}")
    print(f"공격력: {my_pikachu['attack']}")
    print(f"HP: {my_pikachu['hp']}")
    print(f"방어력: {my_pikachu['defense']}")
else:
    print("피카츄를 찾지 못했어요 😢")
```

```python
# 피카츄를 찾았다면 그래프 그리기
if len(pikachu) > 0:
    pikachu_speed = my_pikachu['speed']
    pikachu_attack = my_pikachu['attack']
    
    plt.figure(figsize=(10, 8))
    plt.plot(pikachu_speed, pikachu_attack, 'yo', markersize=25)
    plt.text(pikachu_speed + 3, pikachu_attack + 3, ' 피카츄', fontsize=16)
    
    plt.grid(True, alpha=0.3)
    plt.xlabel('스피드', fontsize=14)
    plt.ylabel('공격력', fontsize=14)
    plt.title(' 피카츄의 스피드 vs 공격력', fontsize=16)
    
    # 축 범위 설정 (피카츄 중심으로)
    plt.xlim(0, pikachu_speed + 20)
    plt.ylim(0, pikachu_attack + 20)
    
    plt.show()
    
    print(f" 피카츄는 스피드 {pikachu_speed}, 공격력 {pikachu_attack}이에요!")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 피카츄 능력치 (위에서 확인한 데이터)
pikachu_speed = 90
pikachu_attack = 55

print(f" 피카츄 능력치: 스피드 {pikachu_speed}, 공격력 {pikachu_attack}")

# 벡터(화살표)로 그리기
plt.figure(figsize=(10, 8))
plt.arrow(0, 0, pikachu_speed, pikachu_attack, 
          head_width=3, head_length=4, 
          fc='yellow', ec='orange', linewidth=3)

plt.text(pikachu_speed + 5, pikachu_attack + 5, 
         '⚡ 피카츄', fontsize=16,
         bbox=dict(boxstyle="round,pad=0.5", facecolor="lightyellow"))

plt.grid(True, alpha=0.3)
plt.xlabel('스피드 →', fontsize=14)
plt.ylabel('공격력 ↑', fontsize=14) 
plt.title('⚡ 피카츄 능력치 벡터', fontsize=16)

plt.xlim(0, pikachu_speed + 20)
plt.ylim(0, pikachu_attack + 20)

plt.show()

# 벡터 길이 계산
vector_length = np.sqrt(pikachu_speed**2 + pikachu_attack**2)
print(f"📏 피카츄 벡터 길이: {vector_length:.2f}")
```
